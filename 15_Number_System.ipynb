        <ul><h3>Babylonians ka Base-60 System (circa 2000 B.C.)</h3>
        <li>Babylonian civilization ne pehli martaba ek structured number system develop kiya, jise Base-60 system kehte hain. </li><br>
            <li>Yeh system kuch specific symbols ke saath different quantities ko represent karta tha.</li><br>
            <li> Unke system me alag-alag positions ka use kiya jata tha jisme higher ya lower value ko represent kiya jata tha, bilkul aaj ke decimal (Base-10) system ki tarah.  </li><br>
            <li>Is system ka asar aaj bhi hai - jaise ghanton aur minutes ka structure (60 seconds, 60 minutes) Base-60 ki wajha se aya hai.</li><br>
            <li>Exampple Agar Babylonians ko 125 represent karna ho, toh yeh Base-60 mein kuch is tarah convert hoga:<br><br>
            <ul>
                <li>125 / 60 = 2 remainder 5</li>
                <li>Yahaan 2 Base-60 ka first position value ban gaya aur 5 remainder bacha.</li>
                <li>Ab Babylonians ke representation ke hisaab se, yeh 2, 5 likha jaayega.</li>
                <li>Iska matlab yeh hai ke 125 ko represent karne ke liye Babylonians yeh symbols likhte, jo ki 2 (× 60) + 5 ka combination hota hai.</li>
                <li>So, 125 Base-60 system mein 2:5 ban jaata hai.</li>
            </ul>
            </li>
        </ul>
        </div></body>

        <ul><h3>Egyptian Number System (circa 3000 B.C.)</h3>
        <li>Misri (Egyptian) civilization ka number system kaafi unique tha. Unka system hieroglyphs ka use karta tha jo alag-alag symbols me numbers ko represent karta tha.</li><br>
            <li>Ye system decimal-based tha (Base-10), lekin unke paas positional value ka concept nahi tha, jise hum aaj ke time me units, tens, aur hundreds ke form me dekhte hain. </li><br>
            <li>Yeh sirf 1, 10, 100, aur 1000 ke multiples ko represent karne ke liye symbols ka use karte the.</li><br>
            <li>Egyptian system mein kuch basic symbols the jo har digit ke liye alag alag cheezein represent karte the:<br><br>
            <ul>
                <li>1 ke liye ek | vertical line</li>
                <li>10 ke liye ek ∩ (boomerang ki tarah)</li>
                <li>100 ke liye ek coil ki tarah ka symbol</li>
                <li>1,000 ke liye ek lotus flower</li>
                <li>10,000 ke liye ek pointing finger</li>
                <li>100,000 ke liye ek frog</li>
                <li>1,000,000 ke liye ek person with raised hands</li>
            </ul>
            </li><br>
            <li>Example: Egyptian Number System mein 276 kaise likhte</li><br>
            <li>200 ke liye: 2 coils banaayein kyunki har coil 100 represent karta hai.</li>
            <li>70 ke liye: 7 boomerangs (∩) banaayein kyunki har boomerang 10 represent karta hai.</li>
            <li>6 ke liye: 6 vertical lines (|) banaayein kyunki har line 1 represent karti hai.</li><br>
            <li>Toh, 276 ko hum is tarah likhte:<br><br>
            <ul>
                <li>2 × (100) ke liye: ◉ ◉ (2 coils)</li>
                <li>7 × (10) ke liye: ∩ ∩ ∩ ∩ ∩ ∩ ∩ (7 boomerangs)</li>
                <li>6 × (1) ke liye: | | | | | | (6 vertical lines)</li>
            </ul>
            </li><br>
            <li>276 = ◉ ◉   ∩ ∩ ∩ ∩ ∩ ∩ ∩   | | | | | |</li>
        </ul>

        <li>Roman civilization ne apna khud ka number system banaya jise Roman Numerals kehte hain. Isme I, V, X, L, C, D, aur M letters ka istemal hota tha jo alag-alag values represent karte the.</li><br>

        <li>Indian mathematicians ne zero ka concept invent kiya aur place value yaani positional notation ko introduce kiya. </li><br>
            <li>Yeh wo breakthrough tha jisne modern number system ka foundation rakha. India me jo system use hota tha woh Base-10 par tha aur isme 0 se 9 tak ke digits ka use kiya jata tha. </li><br>
            <li>Yeh system baad me Arabic scholars ke through Europe tak pahuncha, jise hum aaj Hindu-Arabic numeral system kehte hain. Yeh system calculations ke liye efficient aur easy tha aur iske spread hone se scientific aur mathematical advancements me kaafi help mili.</li><br>
            <li>Example: 437 ko base-10 mein samajhne ka matlab hai:</li><br>
            <li>4 × (10^2) = 400 +</li>
            <li>3 × (10^1) = 30 +</li>
            <li>7 × (10^0) = 7</li>

            <li>Octal ke har digit ko 3 binary digits (bits) mein convert kiya ja sakta hai. Yeh 1 octal digit ko 3 binary bits ke sequence mein represent karta hai</li><br>
