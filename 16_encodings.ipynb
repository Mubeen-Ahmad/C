<h1 style="text-align:center; color:#005bbd; font-size:30px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px Black;">
  Encoding
</h1>

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <ul>
        <li>Encoding aik tarika hai jis se computer non-numeric data ko bhi samajh sakta hai aur store aur process kar sakta hai.</li> <br>
            <li>Computer ke liye sab kuch binary me hota hai, yani 0 aur 1. Lekin jab human-readable text, jaise ke letters aur symbols, ko store karna ho, toh binary format me usko represent karne ka aik systematic way chahiye hota hai.</li> <br>
            <li>Encoding se ye problem solve hui ke har language aur symbol ko binary me convert kiya ja sake aur wapas samjha bhi ja sake.</li>
        </ul>
        <p>Jab hum kisi character ko computer me store karte hain, toh us character ko aik specific number assign kar dete hain. Phir ye number binary me convert hota hai aur computer me store ho jata hai. Jab ye binary data wapas read hota hai, toh encoding system us number ko dobara readable character me convert kar deta hai.</p>
</div></body>



<h1 style="text-align:center; color:#005bbd; font-size:30px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px Black;">
  Story of Encoding
</h1>

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <p>Computer mein characters ko store karne ke encoding systems, jaise ASCII aur Unicode, ki kahani computer science ke initial development se shuru hoti hai. Yeh ek aise waqt mein aayi jab machines ko characters aur numbers ko samajhne aur store karne ki zarurat thi</p>
</div></body>



<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <h2>Telegraph</h2>
        <ul>
            <li>Telegraph aik purana communication system tha jo 1830s ke dauran invent hua. <br>
            <ul>
                <h3>Machine aur Wires:</h3> 
                <li>Telegraph system mein sender aur receiver ke paas telegraph machines hoti thin, jo ek doosre se wires ke zariye connect hoti thi. Yeh wires ke madhyam se signals ko transmit karti thin.</li><br>
             <h3>Electrical Signals</h3>
                <li>Messages ko bhejne ke liye machine se chhoti aur lambi electrical pulses bheje jate. Ye pulses line mein travel karke receiver machine tak pohonchte, jahan unhein decode kiya jata.</li><br>
                <h3>Button Key</h3>
                <li>Telegraph machine mein aik button hota tha, jo "key" kehlaata tha. Sender button ko press karke chhoti aur lambi pulses bhejta:<ul><br>
                    <li>Agar sender button ko sirf aik second ke liye press karta, to yeh short signal yaani "dot" represent karta.</li><br>
                    <li>Agar button ko thodi lambi der ke liye press kiya jata, to yeh long signal yaani "dash" represent karta.</li>
                </ul></li>
        <h3>Receiving aur Decoding</h3>
                <li>Receiver side par machine par yeh electrical signals sunayi deti thi ya ek indicator mein dikhayi deti thi. Receiver unhein decode karta aur samajhta tha ke kaunsa letter ya word bheja gaya hai.</li><br>
        <h3>Example</h3>
                <li>Agar sender ne kuch transmit karna ho, jaise "H," to wo machine ka button chaar baar Thori der ke liye press karega. Yeh chaar "dots" banayenge jo ke "H" ka Morse Code hai.</li><br>
                <li>Telegraph sirf dots aur dashes ke patterns mein hi signals ko bhejne aur receive karne ke liye design ki gayi thi, aur is wajah se yeh long-distance communication ke liye uss waqt kaafi effective aur efficient solution tha.</li>
            </ul>
            </li>
        </ul>
</div></body>



<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <h2>Morse Code</h2>
        <ul>
            <li>Telegraph system jab pehli baar invent hua to usmein messages ko transmit karna aasan nahi tha, kyunke koi standard code ya language nahi thi jo easily signals ko letters ya words mein convert kar sake. Yahi wajah thi ke Samuel Morse ne Morse Code ko invent kiya. <br>
            <ul>
                <h3>Background</h3> 
                <li>1830s ke daur mein Samuel Morse aur unke saathiyon ne ye dekha ke agar electrical signals ko short aur long pulses ke patterns mein bheja jaye, to usay kuch standard letters aur numbers mein convert kiya ja sakta hai.</li><br>
                <li>Iske liye unhone har letter aur number ko ek unique pattern diya. Yeh patterns dots (.) aur dashes (-) mein define kiye gaye.</li>
             <h3>Idea of Coding</h3>
                <li>Morse aur unke team ne socha ke agar har letter ko chhoti aur lambi electrical signals ke patterns mein represent kar diya jaye, to telegraph ke zariye messages bhejne ka tareeqa kaafi aasan ho jayega.</li><br>
                <h3>Standard in Morse Code</h3>
                <li>Morse aur unke team ne alphabet aur numbers ke liye unique patterns design kiye jo dots aur dashes mein thay. Example ke liye:<ul><br>
                    <li>"A" ko .- diya gaya, jisme aik dot aur ek dash tha.</li><br>
                    <li>"B" ko -... diya gaya, jisme aik dash aur teen dots thay.</li>
                </ul></li>
        <h3>Message Encoding</h3>
                <li>Morse Code ke aane se telegraph operators ke liye messages ko encode aur decode karna asaan ho gaya. Ab wo sirf dots aur dashes ke patterns ko padhte aur messages ko easily samajh sakte thay.</li><br>
        <h3>First Demonstration</h3>
                <li>Samuel Morse ne 1844 mein pehli baar Morse Code ka istemal karke Washington, D.C. aur Baltimore ke darmiyan pehla message successfully transmit kiya. Pehla message tha: "What hath God wrought."</li><br>
            </ul>
            </li>
        </ul>
</div></body>

        <h2>Punch Cards</h2>
        <ul>
            <li>Punch cards ka concept sabse pehle Herman Hollerith ne 1890s mein develop kiya tha. Yeh aik hard paper card hoti thi jisme specific jagaon par chhote holes punch kiye jate thay. Har hole ya no-hole, information ko represent karta tha. Yeh approach tab ki machinery aur computers ke liye kaafi innovative thi.<br><br>
            <ul>
                <li>Example: Agar hum chahte hain ke punch card par number "101" ko represent karein, toh hum kuch is tarah se holes punch karenge:<ul><br>
                    <li>1 (Hole): Punch card par hole banaya gaya.</li>
                    <li>0 (No-Hole): Punch card par hole nahi banaya gaya.</li>
                    <li>1 (Hole): Dobara hole banaya gaya.</li>
                </ul>
                </li><br>
                <li>Is tarah, aik punch card series me holes aur no-holes ka combination aik data set represent karta hai jo tab ki machines ya phir industrial computers ko samajh aata tha.</li><br>
        </ul>
</div></body>



<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <h2>Binary Codes</h2>
        <ul>
            <li>1930s ke dauraan binary number system ka istemal computing ke liye zyada taraha se samajh mein aane laga, jo numbers aur letters ko represent karne mein asaani provide karta tha. George Boole ne binary logic ka concept diya, jise Boolean logic kaha jaata hai. Yeh hi wo logic hai jise computer aaj bhi use karta hai, aur jisme 1 aur 0 ke form mein information store hoti hai.</li><br>
            <li> Binary data sirf do states ko represent karta hai: 0 (off) aur 1 (on). Yeh concept electricity ke basis par kaam karta hai aur sirf 2 signals ya phases ka istamal hota hai, yani kuch bhi represent karne ke liye, jo bhi data ho 0 aur 1 ka combination banaye ga.<br><br>
            <ul>
                <li>Example: Agar hum "A" letter ko binary code me likhna chahein, toh ASCII me iska binary code 01000001 hota hai. Yeh binary numbers computer ko samajh aate hain aur woh is code ko A me convert kar leta hai.<ul><br>
                </ul>
                </li><br>
        </ul>
                Punch cards aur binary codes ke darmiyan ek deep connection hai.<br>Punch cards binary ke basic concept ko physical form me represent karti thi — jahan ek hole ko 1 aur no-hole ko 0 samjha jaata tha.<br><br>Lekin punch cards ka ye tareeqa direct binary representation nahi tha; balke yeh aik physical medium tha jo binary logic ke sath kaam karta tha.
</div></body>



<h1 style="text-align:center; color:#005bbd; font-size:30px; font-family:Sans-serif; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px Black;">
  Modern Encodings
</h1>

<body style="font-family: Sans-serif;">
    <div style="color: black; font-size: 18px; font-style: oblique; text-shadow: 0 0 3px white, 0 0 1px black; padding: 20px;">
        <h2>ASCII ka Invention</h2>
        <ul>
        <li>ASCII (American Standard Code for Information Interchange) aik character encoding standard hai jo 1960s mein develop kiya gaya.</li><br>
            <li> ASCII ka invention 1963 mein American National Standards Institute (ANSI) ne develop kiya aur 1968 mein officially release hua.</li><br>
        <li>ASCII ko Bob Bemer ke efforts ki wajah se banaya gaya tha, jo tab IBM mein aik engineer the. Bob Bemer ne dekha ke us waqt ke alag alag computers aur systems ke beech data aur text ka communication mushkil tha kyunki har system apna alag encoding system use karta tha. </li><br>
            <li>Bob Bemer ne propose kiya ke aik universal standard hona chahiye jo text aur special characters ko represent kar sake taake har system aur computer asaani se communicate kar sakein.</li><br>
        <li>ASCII aik 7-bit code ka system hai jo 128 characters ko represent karta hai, jisme English alphabets (a-z aur A-Z), digits (0-9) aur kuch special symbols aur control codes shamil hain.</li><br>
            <h3>ASCII Ka Basic Concept</h3>
            <ul>
                <li>ASCII ka main idea yeh tha ke har character ya symbol ko aik unique number (binary code) assign kiya jaaye jo universally recognized ho. </li><br>
                <li>Aise, computer characters ko binary mein samajh sakta hai aur display kar sakta hai.</li>
            </ul>
        <h3>ASCII Codes Structure</h3>
            <ul>
                <li>ASCII code ko 7-bits mein represent kiya jata hai, jo 2^7 = 128 combinations possible banata hai. In 128 characters ko hum categories mein divide kar sakte hain:<br><br>
                <ul><h3>Control Characters (0-31 aur 127):</h3>
                    <li>Yeh woh codes hain jo printing ke liye nahi hote balkay different actions perform karne ke liye use hote hain, jaise ke newline, tab, etc.</li><br>
                </ul>
                    <ul><h3>Printable Characters (32-126):</h3>
                    <li>Inme woh symbols hain jo screen par show ho sakte hain, jisme letters, digits aur punctuation shamil hain.</li>
                </ul>
                </li>
        </ul>
        <h3>Examples</h3>
        <ul>ASCII mein English alphabets ko specific numbers se map kiya gaya hai.<br><br>
            <li>Capital "A" ka ASCII code 65 hota hai, jo binary mein 1000001 hai.</li>
            <li>Small "a" ka ASCII code 97 hai, jo binary mein 1100001 hai.</ul></li>
        </ul>

        <h2>Unicode</h2><br>
        <ul><li>Unicode ka aana 1990s mein hua, jab computers duniya bhar mein use honay lage aur alag alag languages aur symbols ko represent karna zaroori ho gaya.</li><br>
            <li>ASCII aur Extended ASCII ke limitations ki wajah se in systems mein bas English aur kuch special symbols hi represent kiye ja sakte the. Lekin jab global languages, jaise Chinese, Arabic, Hindi, aur special symbols ki zaroorat hui, to Unicode standard introduce kiya gaya.</li><br>
            <li>Unicode ne har language aur special symbol ke liye ek unique code provide kiya, jise sab computers aur systems samajh sakte hain. Unicode ka ye standard Unicode Consortium ne 1991 mein establish kiya.</li><br>
    <li>Unicode mein alag alag encoding systems aaye jo har character ko represent karne ke liye multiple bytes ka use karte hain. Kuch common encodings hain:<br><br>
    <ul>
        <h3>UTF-8</h3>
        <li>Ye aaj ka sabse common encoding system hai aur backward compatible bhi hai ASCII ke saath.</li><br>
        <li>UTF-8 mein characters 1 se lekar 4 bytes tak ke space mein store kiye ja sakte hain.</li>
        <li>Iska fayda ye hai ke ye English characters ke liye bas 1 byte use karta hai (ASCII ke jese), lekin complex symbols aur global languages ke liye 2, 3, ya 4 bytes tak expand ho sakta hai.</li><br>
        <h3>UTF-16</h3>
        <li>UTF-16 2 ya 4 bytes ka use karta hai.</li><br>
        <li>Ye mainly jab high-frequency characters aur kuch specific complex languages use hoti hain, to 2 ya 4 bytes me expand ho sakta hai.</li><br>
<h3>UTF-32</h3>
        <li>UTF-32 4 bytes ka fixed space use karta hai aur har character ko 4 bytes mein store karta hai.</li><br>
        <li>Iski wajah se ye memory mein zyada jagah leta hai, lekin iski encoding simple hai.</li><br>
    </ul>
    </li>
    <h3>Examples</h3>
            <li>ASCII aur Unicode dono mein "A" ka code 65 hota hai.</li>
            <li>UTF-8 mein "A" bas 1 byte mein store hota hai (same as ASCII): 01000001</li><br>
            <li>Chinese Character "你" (You):
                <ul>
                    <li>ASCII aur Extended ASCII mein ye character nahi aa sakta, lekin Unicode mein iske liye ek unique code hai.</li>
                    <li>Unicode mein "你" ka code 20320 hai.</li>
                    <li>UTF-8 mein ye 3 bytes mein store hota hai: E4 BD A0</li>
                    <li>UTF-16 mein ye 2 bytes mein store hota hai: 4F60</li>
                </ul>
            </li>
        </ul>
<p>Unicode ka ek specific table nahi hota, kyunki Unicode ka range aur character set bohot zyada bada hai. Unicode mein 1.4 million se zyada code points hain jo har tarah ke languages aur symbols ko cover karte hain.</p>
